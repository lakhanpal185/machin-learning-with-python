```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression  
import matplotlib.pyplot as plt
import seaborn as sns


iris_data = pd.read_csv('/content/drive/MyDrive/homeprices (1).csv')
#print(iris_data.head())
#print(iris_data.shape)

#df_binary = [['area','price']]
sns.lmplot(x = 'area', y = 'price',data = iris_data,order = 2,ci = None)
iris_data.dropna( inplace = True)

x_train = np.array(iris_data['area']).reshape(-1,1)
print(x_train.shape)
regr = LinearRegression();


regr.fit(x_train,iris_data['price'])
print(regr.coef_)
print(regr.intercept_)
y_predi = regr.predict([[1000]])
print(f"the y perdicted is {y_predi}")



newdata = pd.read_csv('/content/drive/MyDrive/myarea.csv')
print(newdata)
predictions = regr.predict(newdata)
newdata['price'] = predictions

newdata.to_csv("mypredictions.csv")
```
