```python
import seaborn as sns
data = sns.load_dataset('tips')
print(data.to_string())
```

```python
sns.boxplot(x = data['total_bill'],y = data['sex'])
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = data,)
```

```python
sns.boxplot(x = 'day',y ='total_bill',hue ='smoker',data = data,palette = 'Set2')
```
