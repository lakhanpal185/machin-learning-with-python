```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import linear_model
from sklearn.metrics import classification_report,accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/insurance_data.csv')
df.rename(columns = {'bought_insurance':'get_insur'},inplace = True)
df
```

```python
sns.regplot(x = "age",
            y = "get_insur", 
            ci = None,
            data = df)
```

```python
df.drop_duplicates(inplace = True)
df
```

```python
lrg = linear_model.LogisticRegression()
lrg.fit(df[['age']],df['get_insur'])

y_pred = lrg.predict(df[['age']])
y_pred

print(accuracy_score(df['get_insur'],y_pred))
```
