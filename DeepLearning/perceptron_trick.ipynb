```python
from sklearn.datasets import make_classification
import numpy as np
X , y = make_classification(n_samples=1000, n_features=2, n_informative=1, n_classes=2, n_clusters_per_class=1,
                            n_redundant=0, random_state=41, hypercube=False, class_sep = 10 )
print(X.shape, y.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,5))
plt.scatter(X[:,0], X[:,1], c = y,cmap = 'winter',s =5)
print(y[0])
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis =1)
    w = np.zeros(X.shape[1])
    lr = 0.01
    for i in range(100000):
        j = np.random.randint(0,1000)
        y_pred = step(np.dot(X[j],w))
        w = w + lr*(y[j] - y_pred)*X[j]

    return w[0],w[1:]
perceptron(X,y)
```

```python
def myperceptron(X,y):
    X = np.insert(X,0,1,axis =1)
    w = np.ones(X.shape[1])
    for i in range(300):
        y_pred =  np.dot(X,w)
        for i in range(len(y_pred)):
            if y_pred[i] > 0:
                y_pred[i] = 1
            else:
                y_pred[i] = 0
            if y_pred[i] != y[i]:
                w = w + 0.1*y[i]*X[i]
            else :
                print(y_pred[i],y[i])
    return w[0],w[1:]
myperceptron(X,y)
```

```python
def step(x):
    return 1 if x > 0 else 0
```

```python
intercept_,coef_ = myperceptron(X,y)
```

```python
m = - coef_[0]/coef_[1]
b = - intercept_/coef_[1]
```
