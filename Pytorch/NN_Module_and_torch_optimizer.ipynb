# Using layers from NN Module

```python
import torch.nn as nn
import torch
```

```python
class Model(nn.Module):
    def __init__(self,num_features):
        super().__init__()
        self.layer_1 = nn.Linear(num_features,1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, data):
        out = self.layer_1(data)
        out = self.sigmoid(out)
        return out
```

```python
# create dataset
dataset = torch.rand(10,5)
```

```python
#creating the model
model = Model(dataset.shape[1])
```

```python
#doing forward pass
model(dataset)
```

# Training using NN module components
* Optimizer
* Loss function
* Layers

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder


#data preparation
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
df.head()
```

```python
X_train,X_test, y_train, y_test= train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

# Designing the model architecture

```python
class Model(nn.Module):
    def __init__(self,num_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_features,3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )

    def forward(self, data):
        out = self.network(data)

        return out
```

```python
learning_rate = 0.01
num_epochs = 100

loss_fn = nn.BCELoss()
```

# Training


```python
model = Model(X_train_tensor.shape[1])

optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)

for epoch in range(num_epochs):

    y_pred = model(X_train_tensor)

    loss = loss_fn(y_pred.squeeze(1),y_train_tensor)
    print(f'Epoch: {epoch} Loss: {loss.item()}')
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

```

```python
#Evalution
from sklearn.metrics import accuracy_score


with torch.no_grad():
    y_pred = model(X_test_tensor)
    y_pred_labels = torch.round(y_pred)
    acc = accuracy_score(y_test_tensor, y_pred_labels)
    print(f'Accuracy: {acc}')
```
